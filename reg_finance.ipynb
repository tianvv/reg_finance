```python
import numpy as np
import matplotlib as mpl
import matplotlib.cm as cm
import matplotlib.pyplot as plt
import pandas as pd
from pandas.tseries.offsets import MonthEnd
```

```python
import seaborn as sns
import sklearn as sk
from sklearn.decomposition import PCA
from sklearn import preprocessing
import datetime
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
df_finance= pd.read_csv('finance2.csv')
df_finance=df_finance.dropna(axis=0,how='any')
```

```python
stkcd_df_feature =pd.DataFrame(df_finance.groupby('Stkcd').size())
stkcd_array_feature=stkcd_df_feature.index[stkcd_df_feature[0] == 38].values
stkcd_array_feature
```

```python
stkcd_list = []

for i in range(max(stkcd_array_feature)):
    if i in stkcd_array_feature:
        stkcd_list.append(i)


df_finance= df_finance[df_finance['Stkcd'].isin(stkcd_list)]
```

```python
df_finance1=df_finance.copy()
df_label=df_finance1.loc[:,['Stkcd','Accper','Mretwd','rf']]
df_feature=df_finance1.loc[:,['F1','F2','F3','F4','F5','F6','F7','F8','F9','F10','F11','F12','F13','F14','F15','F16','F17','F18']]
```

```python
def zscore(group):
    return(group-group.mean())/group.std()
```

```python
df_std=df_feature.apply(zscore)
df_std
```

```python
df_label['rirf']=df_label['Mretwd']-df_label['rf']
df_label
```

```python
df_label.to_csv('ret.csv',index=False)
```

```python
df_finance=pd.concat([df_label,df_std],axis=1)
```

```python
df_train=pd.DataFrame()
df_train=df_finance[df_finance['Accper']!='09/01/17']
```

```python
df_test=df_finance[df_finance['Accper']=='09/01/17']
```

```python
def regress(data,y,x):
    Y=data[y]
    X=data[x]
    result=sm.OLS(Y,X).fit()
    return result.params
```

```python
betas=df_train.groupby(['Stkcd','Accper']).apply(regress,'rirf',['F1','F2','F3','F4','F5','F6','F7','F8','F9','F10','F11','F12','F13','F14','F15','F16','F17','F18'])
```

```python
beta=betas.groupby(['Stkcd']).mean()
```

```python

df_test=df_test.groupby(['Stkcd']).mean()
df_test_feature=df_test.loc[:,['F1','F2','F3','F4','F5','F6','F7','F8','F9','F10','F11','F12','F13','F14','F15','F16','F17','F18']]
```

```python
retm=np.dot(beta,df_test_feature.T)
```

```python
ret=pd.Series([retm[i][i] for i in range(len(retm))])
ret
```
